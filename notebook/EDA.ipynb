```python
def cleandata(df):
    from sklearn.preprocessing import LabelEncoder
    import pandas as pd

    df = df.copy()  # Prevent modifying the original DataFrame

    # Convert datetime and timestamp properly
    df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')  
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s", errors='coerce')

    # Create duration feature (distance / avg_speed * 60)
    df['duration'] = (df['distance'] / 20) * 60  

    # Extract datetime features
    df['hour'] = df['datetime'].dt.hour
    df['day'] = df['datetime'].dt.day
    df['month'] = df['datetime'].dt.month

    # Define required columns
    column = ['datetime', 'hour', 'day', 'month', 
              'source', 'destination', 'cab_type', 'name', 
              'price', 'distance', 'surge_multiplier', 
              'apparentTemperature', 'short_summary', 
              'precipIntensity', 'precipProbability', 'icon',
              'latitude','longitude','duration']
    
    # Keep only relevant columns
    df = df[[col for col in column if col in df.columns]]
    
    # Label encode categorical features
    label_encoder = LabelEncoder()
    for col in ["source", "destination", "cab_type", "name", "short_summary", "icon"]:
        if col in df.columns:
            df[col + "_encoded"] = label_encoder.fit_transform(df[col].astype(str))

    # Drop rows with missing price
    df = df.dropna(subset=["price"])

    return df
```

```python
df=cleandata(data)
df.isnull().sum()
```

```python
# the distibution of rides per differrent month based on lyft or uber
plt.figure(figsize=(8,5))
sns.countplot(x=df['month'], hue=df['cab_type'], palette='viridis')

plt.title('Distribution of rides per month')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['hour'], hue=df['cab_type'], palette='viridis')

plt.title('Distribution of rides per hour')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['cab_type'], hue=df['name'], palette='viridis')

plt.title('Distribution of rides per hour')
plt.show()
```

```python
# Set the overall plot size
plt.figure(figsize=(16, 8))

# Uber price distribution
plt.subplot(1, 2, 1)
sns.violinplot(data=df[df['cab_type'] == 'Uber'], x='name', y='price', palette='viridis', inner="quart", linewidth=1.2)
plt.title('Price Distribution for Uber Rides by Class')
plt.xlabel('Ride Type')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)

# Lyft price distribution
plt.subplot(1, 2, 2)
sns.violinplot(data=df[df['cab_type'] == 'Lyft'], x='name', y='price', palette='viridis', inner="quart", linewidth=1.2)
plt.title('Price Distribution for Lyft Rides by Class')
plt.xlabel('Ride Type')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout();

plt.show();
```
